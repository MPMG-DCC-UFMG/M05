```python
import pandas as pd
```

```python
original_csv_filepath = '../consumidorgovbr.csv'
new_csv_filepath = '../consumidor_gov.csv.gz'
```

```python
df_consumidor = pd.read_csv(original_csv_filepath, low_memory=False)
```

```python
df_consumidor.columns
```

```python
df_consumidor['titulo'].replace(['\n\t\t\t\t'], '- ', inplace=True, regex=True)
```

```python
df_consumidor.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
def adjust_date_format(value):
    day, month, year = value.strip().split('/')
    return year+'-'+month+'-'+day
```

```python
df_consumidor['data_criacao'] = df_consumidor['data_criacao'].apply(adjust_date_format)
```

```python
df_consumidor
```

```python
df_consumidor[:100000].to_csv('../consumidor_gov_01.csv.gz', index=False, compression='gzip')
df_consumidor[100000:200000].to_csv('../consumidor_gov_02.csv.gz', index=False, compression='gzip')
df_consumidor[200000:300000].to_csv('../consumidor_gov_03.csv.gz', index=False, compression='gzip')
df_consumidor[300000:400000].to_csv('../consumidor_gov_04.csv.gz', index=False, compression='gzip')
df_consumidor[400000:500000].to_csv('../consumidor_gov_05.csv.gz', index=False, compression='gzip')
df_consumidor[500000:600000].to_csv('../consumidor_gov_06.csv.gz', index=False, compression='gzip')
df_consumidor[600000:700000].to_csv('../consumidor_gov_07.csv.gz', index=False, compression='gzip')
df_consumidor[700000:].to_csv('../consumidor_gov_08.csv.gz', index=False, compression='gzip')
```
