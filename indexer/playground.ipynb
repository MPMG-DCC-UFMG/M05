```python
import numpy as np
import pandas as pd
import csv
import json
import os
```

```python
def generate_formated_csv_lines(file_path, index_name, doc_type, encoding="utf8"):
    file = open(file_path, encoding=encoding)
    table = csv.DictReader(file)
    columns = table.fieldnames.copy()
    
    for line in table:
        line = dict(line)
        doc = {}
        for field in columns:
            doc[field] = line[field]
        
        doc["fonte_csv"] = file_path
    
        yield {
            "_index": index_name,
            "_type": doc_type,
            "_source": doc
        }
```

```python
file = "csvs/73-diarios-governador_valadares.csv"
```

```python
for line in generate_formated_csv_lines(file, "diarios", "diario"):
    print(line["_source"])
    break
```
