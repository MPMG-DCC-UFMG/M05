```python
from os import listdir
from os.path import isfile, join

def list_files(mypath):
    if mypath[-1] != "/":
        mypath = mypath+"/"
    return [mypath+f for f in listdir(mypath) if isfile(join(mypath, f))]

def list_dirs(folder):
    return [ f.path for f in os.scandir(folder) if f.is_dir() ]
```

# Treating 37-MG

```python
errors = []
n_files = 0

csv_file = open("csvs/37-diario-oficial-mg.csv", 'w')
csv_writer = csv.DictWriter(csv_file, ["conteudo", "fonte", "data"])
csv_writer.writeheader()

for ano in list_dirs("original/37-diario-oficial-mg/"):
    for file in list_files(ano):
        
        n_files += 1
        
        try:
            csv_writer.writerow({"conteudo": parser.from_file(file, requestOptions={'timeout': 300})["content"],
                                 "fonte": file,
                                 "data": "-".join(file.split("/")[-1][:10].split("-")[::-1]) })
        except:
            errors.append(file)

csv_file.close()
            
print(errors)
print(n_files)
```
