```python
import pandas as pd
```

```python
original_csv_filepath = '../consumidorgovbr.csv'
new_csv_filepath = '../consumidor_gov.csv'
```

```python
df_consumidor = pd.read_csv(original_csv_filepath, low_memory=False)
```

```python
df_consumidor.columns
```

```python
df_consumidor['titulo'].replace(['\n\t\t\t\t'], '- ', inplace=True, regex=True)
```

```python
df_consumidor.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
def adjust_date_format(value):
    day, month, year = value.strip().split('/')
    return year+'-'+month+'-'+day
```

```python
df_consumidor['data_criacao'] = df_consumidor['data_criacao'].apply(adjust_date_format)
```

```python
df_sample = df_consumidor[:1000].copy()
```

```python
df_sample
```

```python
df_sample.to_csv(new_csv_filepath, index=False)
```
