```python
import pandas as pd
```

```python
original_csv_filepath = '../procon_original.csv'
new_csv_filepath = '../procon.csv.gz'
```

```python
df_procon = pd.read_csv(original_csv_filepath, low_memory=False)
```

```python
df_procon.columns
```

```python
new_df = df_procon[['id_manifestacao', 'data_manifestacao_show', 'nome_pessoa_show', 'titulo', 'mensagem_show', 'nome_completo_empresa', 'nome_cidade_ocorrencia_show', 'nom_sub_classe']].copy() #[:5000].copy()
```

```python
new_df.rename(columns={'data_manifestacao_show':'data_criacao', 'nome_pessoa_show':'nome_autor', 'mensagem_show':'conteudo', 'nome_cidade_ocorrencia_show':'cidade', 'nom_sub_classe':'categoria'}, inplace=True)
```

```python
new_df
```

```python
new_df.to_csv(new_csv_filepath, index=False, compression='gzip')
```
